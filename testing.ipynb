```python
from tensorflow.keras.preprocessing.image import img_to_array, load_img
import matplotlib.pyplot as plt
from keras.models import load_model
import numpy as np
import warnings
warnings.filterwarnings('ignore')

# Get the prediction
img = "/home/sky/Desktop/driver_drowsiness/dataset/test/1/24.jpg"
img = load_img(img, target_size=(128, 128))
plt.imshow(img)
img = img_to_array(img)
img = np.expand_dims(img, axis=0)

#Load model
model = load_model("/home/sky/Desktop/driver_drowsiness/cnn_m_dd.h5")

class_ = np.argmax(model.predict(img,verbose=0))

if class_ == 0:
    print("PREDICTIONS : level 0")
elif class_ == 1:
    print("PREDICTIONS : level 1")
elif class_ == 2:
    print("PREDICTIONS : level 2")
else:
    print("PREDICTIONS : level 3")

    
```
